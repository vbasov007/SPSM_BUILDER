```python
import pandas as pd
```

```python
df = pd.DataFrame({'B': [1, 2, 3], 'C': [4, 5, 6], 'D': [7, 8, 9], 'E': [10, 11, 12]})
```

```python
df
```

```python
def reorder(df, new_order):
    for pos, c in enumerate(new_order):
        col = df[c].copy()
        df.drop(c, 1, inplace = True)
        df.insert(pos, c, col)
        
```

```python
reorder(df, ['D', 'E', 'C', 'B'])
```
